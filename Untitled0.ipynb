```python
# Importamos las librer√≠as necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# 1. NumPy: crear y manipular arrays
datos = np.array([[1, 2], [3, 4], [5, 6]])
print("Array de NumPy:")
print(datos)
print("Media de cada columna:", np.mean(datos, axis=0))

# 2. Pandas: crear y visualizar un DataFrame
df = pd.DataFrame({
    'Horas_Estudio': [1, 2, 3, 4, 5],
    'Calificaci√≥n': [2, 4, 5, 4, 5]
})
print("\nDataFrame con Pandas:")
print(df)

# 3. Matplotlib y Seaborn: visualizaci√≥n de datos
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Horas_Estudio', y='Calificaci√≥n', data=df)
plt.title("Relaci√≥n entre horas de estudio y calificaci√≥n")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificaci√≥n")
plt.grid(True)
plt.show()

# 4. Scikit-learn: modelo de regresi√≥n lineal
X = df[['Horas_Estudio']]  # Variable independiente
y = df['Calificaci√≥n']     # Variable dependiente

modelo = LinearRegression()
modelo.fit(X, y)

# Mostrar pendiente y ordenada al origen
print("\nModelo de regresi√≥n lineal:")
print(f"Pendiente (coeficiente): {modelo.coef_[0]}")
print(f"Intersecci√≥n (ordenada al origen): {modelo.intercept_}")

# Graficar la l√≠nea de regresi√≥n
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Horas_Estudio', y='Calificaci√≥n', data=df)
plt.plot(df['Horas_Estudio'], modelo.predict(X), color='red', label='L√≠nea de regresi√≥n')
plt.legend()
plt.title("Regresi√≥n Lineal: Horas de Estudio vs Calificaci√≥n")
plt.show()
```
