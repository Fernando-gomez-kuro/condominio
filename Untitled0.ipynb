```python
# Importamos las librerías necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# 1. NumPy: crear y manipular arrays
datos = np.array([[1, 2], [3, 4], [5, 6]])
print("Array de NumPy:")
print(datos)
print("Media de cada columna:", np.mean(datos, axis=0))

# 2. Pandas: crear y visualizar un DataFrame
df = pd.DataFrame({
    'Horas_Estudio': [1, 2, 3, 4, 5],
    'Calificación': [2, 4, 5, 4, 5]
})
print("\nDataFrame con Pandas:")
print(df)

# 3. Matplotlib y Seaborn: visualización de datos
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Horas_Estudio', y='Calificación', data=df)
plt.title("Relación entre horas de estudio y calificación")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificación")
plt.grid(True)
plt.show()

# 4. Scikit-learn: modelo de regresión lineal
X = df[['Horas_Estudio']]  # Variable independiente
y = df['Calificación']     # Variable dependiente

modelo = LinearRegression()
modelo.fit(X, y)

# Mostrar pendiente y ordenada al origen
print("\nModelo de regresión lineal:")
print(f"Pendiente (coeficiente): {modelo.coef_[0]}")
print(f"Intersección (ordenada al origen): {modelo.intercept_}")

# Graficar la línea de regresión
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Horas_Estudio', y='Calificación', data=df)
plt.plot(df['Horas_Estudio'], modelo.predict(X), color='red', label='Línea de regresión')
plt.legend()
plt.title("Regresión Lineal: Horas de Estudio vs Calificación")
plt.show()
```
